```python
from scipy.stats import truncnorm

# تعریف پارامترهای توزیع
a, b = 0, 5  # محدوده برش
loc, scale = 2, 1  # میانگین و انحراف معیار

# نمونه‌گیری از توزیع
fluid_loss = truncnorm.rvs(a, b, loc=loc, scale=scale, size=1000)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# تولید داده‌های نمونه
n = 1000
timestamps = [datetime.now() + timedelta(seconds=i) for i in range(n)]
pressure = np.random.normal(3000, 100, n)
flow_rate = np.random.normal(500, 50, n)
viscosity = np.random.normal(10, 1, n)
depth = np.linspace(1000, 5000, n)
fluid_loss_risk = np.random.uniform(0, 1, n)

# ساخت DataFrame
df = pd.DataFrame({
    'timestamp': timestamps,
    'pressure(psi)': pressure,
    'flow_rate(gpm)': flow_rate,
    'viscosity(cP)': viscosity,
    'depth(ft)': depth,
    'fluid_loss_risk(0-1)': fluid_loss_risk
})

# ذخیره به‌صورت CSV
df.to_csv('synthetic_drilling_data.csv', index=False)

# ذخیره به‌صورت Parquet
df.to_parquet('synthetic_drilling_data.parquet', index=False)
```
